**Task 1 — Conceptual Understanding**

1. Difference between "Love" and "love" in NLP

In NLP, "Love" and "love" are treated as different tokens because NLP is case-sensitive. This increases vocabulary size unnecessarily. Converting text to lowercase helps treat them as the same word and improves model performance.

2. What happens if stopwords are not removed?

If stopwords are not removed:
Vocabulary size increases
Model becomes slower
Important words get less weight
Noise increases in text data

3. Two scenarios where removing stopwords is harmful

1. Sentiment analysis
Example: "I am not happy" → removing "not" changes meaning
2. Question answering systems
Example: "What is the capital of India?"
Removing "what" affects sentence meaning

4. Stemming vs Lemmatization

Stemming: Cuts words roughly and May produce invalid words
Example: "playing" → "play"

Lemmatization: Converts to dictionary root and more accurate
Example: "better" → "good"

### Task 2: Build Advanced Preprocessing Function


```python
import re
import string

def preprocess_text(text):
    
    if not isinstance(text, str):
        return []
    
    # lowercase
    text = text.lower()
    
    # removes urls
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # remove email
    text = re.sub(r'\S+@\S+', '', text)
    
    # removes numbers
    text = re.sub(r'\d+', '', text)
    
    # remove repeated characters (soooo -> soo)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # tokenize
    tokens = text.split()
    
    # remove short words except no, not
    tokens = [word for word in tokens if len(word) > 2 or word in ['no','not']]
    
    return tokens
```

### Task 3: Stress Testing 


```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
for sentence in test_sentences:

    tokens = preprocess_text(sentence)
    cleaned = " ".join(tokens)

    print("Original Text:", sentence)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)
    print("-"*50)
```

## Task 4: Token Analysis 

```python
for sentence in test_sentences:

    tokens = preprocess_text(sentence)

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(w) for w in tokens) / total_tokens if tokens else 0

    print("Sentence:", sentence)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Length:", round(avg_length,2))
    print("-"*40)
```

## Task 5: Frequency Analysis

```python
from collections import Counter
all_tokens = []

for sentence in test_sentences:
    all_tokens.extend(preprocess_text(sentence))

counter = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(counter.most_common(10))

print("\nLeast Frequent Words:")
print(counter.most_common()[:-6:-1])
```

## Task 6: Full Pipeline

```python
def full_pipeline(text_list):

    tokens = []
    clean_sentences = []

    for text in text_list:
        t = preprocess_text(text)
        tokens.extend(t)
        clean_sentences.append(" ".join(t))

    return {
        "tokens": tokens,
        "clean_sentences": clean_sentences
    }
result = full_pipeline(test_sentences)
print(result)
```
